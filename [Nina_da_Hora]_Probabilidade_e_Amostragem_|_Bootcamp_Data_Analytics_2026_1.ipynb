```python
import pandas as pd
import numpy as np

# 1. CARREGAMENTO DA BASE DIRECTO DO GITHUB (com o nosso separador campeão)
url = 'https://raw.githubusercontent.com/stephleao/wmc-desafio-streaming-churn/refs/heads/main/clientes.csv'
df = pd.read_csv(url, sep=';')

# Limpeza básica padrão
df = df.drop_duplicates()

# 2. SUGESTÃO DA STEPHANNE: Convertendo a coluna 'cancelou' para Booleano (True/False)
# Se for 1, 'Sim', 'sim' vira True. Caso contrário, False.
df['cancelou'] = df['cancelou'].isin([1, 'Sim', 'sim', 'S'])

# 3. NOVA SEGMENTAÇÃO (Conforme as diretrizes da Karol e os slides da equipe)
# Criamos as regras apenas para quem tem Churn == True (Cancelados)
condicoes = [
    ((df['cancelou'] == True) & (df['tempo_assinatura_meses'] <= 6)),         # Até 6 meses
    ((df['cancelou'] == True) & (df['tempo_assinatura_meses'] >= 7) & (df['tempo_assinatura_meses'] <= 24)), # 7 a 24 meses
    ((df['cancelou'] == True) & (df['tempo_assinatura_meses'] > 24))        # Acima de 24 meses
]

nomes_grupos = ['Até 6 meses', '7 a 24 meses', 'Acima de 24 meses']

# Aplicando a segmentação. Quem não cancelou (False) vira automaticamente 'Ativos'
df['grupo_cancelamento'] = np.select(condicoes, nomes_grupos, default='Ativos')

# 4. VALIDAÇÃO DOS TOTAIS PARA A KAROL
print("=== VALIDAÇÃO DA SEGMENTAÇÃO DE CHURN ===")
print(df['grupo_cancelamento'].value_counts())

print(f"\nTotal Geral de Clientes: {len(df)}")
print(f"Total de Clientes Cancelados: {df['cancelou'].sum()}")
```
